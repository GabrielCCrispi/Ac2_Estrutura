A função deve usar uma pilha para verificar se a sequência de caracteres está corretamente balanceada. A sequência está balanceada se cada tipo de abertura de parêntese, colchete ou chave tem uma correspondente e os pares são corretamente fechados na ordem reversa.

 - Você deve usar as classes Pilha dinâmica e Fila dinâmica ensinadas na disciplina e disponibilizadas juntamente com este notebook. Elas não podem ser modificadas. 

- As únicas operações permitidas para acessar os valores da pilha e fila são, respectivamente, as operações pop e dequeue. 

- Você deve garantir que ao final da função, a pilha e fila passadas como parâmetro estejam intactas, isto é, como os mesmos valores passados como entrada. 

````python
import sys
sys.path.append('utils')

from pilhaDinamica import Pilha
from filaDinamica import Fila

def verificaBalanceamento(fila):
    pilha = Pilha()
    mapeamento = {')': '(', ']': '[', '}': '{'}
    fila_aux = Fila()

    while not fila.isEmpty():
        char = fila.dequeue()
        fila_aux.enqueue(char)

        if char in mapeamento.values():
            pilha.push(char)
        elif char in mapeamento.keys():
            if pilha.isEmpty() or mapeamento[char] != pilha.pop():
                return False

    while not fila_aux.isEmpty():
        fila.enqueue(fila_aux.dequeue())

    return pilha.isEmpty()

# testando a função criada
fila1 = Fila()
fila1.enqueue('{')
fila1.enqueue('[')
fila1.enqueue('(')
fila1.enqueue(')')
fila1.enqueue('(')
fila1.enqueue(')')
fila1.enqueue(']')
fila1.enqueue('}')

fila2 = Fila()
fila2.enqueue('{')
fila2.enqueue('[')
fila2.enqueue('(')
fila2.enqueue(']')
fila2.enqueue(')')
fila2.enqueue('}')

fila3 = Fila()
fila3.enqueue('{')
fila3.enqueue('{')
fila3.enqueue('[')
fila3.enqueue('[')
fila3.enqueue('(')
fila3.enqueue('(')
fila3.enqueue(')')
fila3.enqueue(')')
fila3.enqueue(']')
fila3.enqueue(']')
fila3.enqueue('}')
fila3.enqueue('}')

statusFila1 = verificaBalanceamento(fila1)
statusFila2 = verificaBalanceamento(fila2)
statusFila3 = verificaBalanceamento(fila3)

print('Resultado retornado:')
print('\tFila 1?: ', statusFila1)
print('\tFila 2?: ', statusFila2)
print('\tFila 3?: ', statusFila3)

print('\n' + 20*'-')
print('Resultado esperado:')
print('\tFila 1?: True')
print('\tFila 2?: False')
print('\tFila 3?: True')
````

## Exercício 3 

Modifique a lista duplamente ligada ensinada nessa disciplina e fornecida juntamente com este trabalho, alterando seu nome para **ListaBuble**. 

Os valores que deverão ser usados nessa lista são **strings**. O critério para você ordenar as strings dessa lista é um **número inteiro** que represente a quantidade de palavras. Você deve considerar como palavra, qualquer conjunto de caracteres que esteja agrupado antes ou depois de um espaço em branco. Por exemplo, suponha que a string contida na posição i da lista seja **Eu fui para um shopping em Sorocaba**. A quantidade de palavras é 7. Já, na frase "Centro Universitário Facens" a quantidade de palavras é 3. Portanto, a primeira string deveria ser posicionada antes da segunda na lista.

Altere também método o **inserir** para que em vez de receber um índice com a posição onde o dado deve ser inserido, sempre insere os valores de maneira ordenada. 

- O método de ordenação aplicado deve ser o método **Bolha (buble sort)**. 
  - No algoritmo Buble sort ensinado na disciplina, a cada iteração, o MAIOR número vai sendo movido para o topo da lista. Modifique o algoritmo de forma que, a cada iteração, o MENOR número vá sendo movido para a parte inicial da lista. Exemplo: na ordenação dos números [2,4,5,1,3], a primeira iteração do Buble sort mostrado na disciplina resulta em [2,4,1,3,5]. Na nova versão que você irá implementar, após a primeira iteração a lista deverá ser [1,2,4,5,3]. Neste exercício, os valores não serão números, mas sim string seguindo essa mesma lógica.
  
**Obs. 1.** Não é necessário modificar os métodos **remove** e **busca** da lista duplamente ligada fornecida. Você pode remover esses métodos da classe que você irá implementar. Porém, caso você queira, você pode alterar o método **busca** para auxiliar o método **inserir**. 

**Obs. 2.** Você poderá usar a função **split** para separar as palavras por espaço em branco. Você pode usar também a função **len** para calcular a quantidade de palavras da frase.  

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class Node:
    """
    Noh de uma lista duplamente ligada
    """
    
    def __init__(self, data = None, next = None, prev = None):
        self.data = data
        self.next = next # apontador para o noh da frente
        self.prev = prev # apontador para o noh de tras

class ListaBolha:
    """
    Lista duplamente ligada ordenada por bolha de strings
    """

    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def separa_palavras(self, string):
        """
        Separa uma string em palavras usando espaços em branco.

        Argumentos:
            string (str): A string a ser separada.

        Retorna:
            list: Uma lista contendo as palavras da string.
        """
        return string.split(" ")

    def conta_palavras(self, string):
        """
        Conta a quantidade de palavras em uma string.

        Argumentos:
            string (str): A string a ser analisada.

        Retorna:
            int: A quantidade de palavras na string.
        """
        return len(self.separa_palavras(string))

    def compara_palavras(self, str1, str2):
        """
        Compara duas strings com base na quantidade de palavras.

        Argumentos:
            str1 (str): A primeira string a ser comparada.
            str2 (str): A segunda string a ser comparada.

        Retorna:
            int: Um valor negativo se str1 tiver menos palavras que str2, 
                 zero se as strings tiverem a mesma quantidade de palavras, 
                 ou um valor positivo se str1 tiver mais palavras que str2.
        """
        conta_palavras_str1 = self.conta_palavras(str1)
        conta_palavras_str2 = self.conta_palavras(str2)

        if conta_palavras_str1 < conta_palavras_str2:
            return -1
        elif conta_palavras_str1 == conta_palavras_str2:
            return 0
        else:
            return 1

    def encontraPontoInser(self, string):
        """
        Encontra o ponto de inserção para uma string na lista ordenada.

        Argumentos:
            string (str): A string a ser inserida.

        Retorna:
            int: O índice onde a string deve ser inserida.
        """

        
        if self.isEmpty():
            return 0

        tempNode = self.head
        insertion_index = 0

        while tempNode is not None:
            if self.compara_palavras(string, tempNode.data) <= 0:
                return insertion_index
            else:
                insertion_index += 1
                tempNode = tempNode.next

        return insertion_index

    def isEmpty(self):
        """
        Verifica se a lista está vazia.

        Retorna:
            bool: True se a lista estiver vazia, False caso contrário.
        """
        return self.size == 0


    def insert(self, string):
        """
        Insere uma string na lista ordenada por bolha de strings.

        Argumentos:
            string (str): A string a ser inserida.
        """
        insertion_index = self.encontraPontoInser(string)
        self.insert_at_index(string, insertion_index)

    def insert_at_index(self, string, index):
        """
        Insere uma string na lista em um determinado índice.

        Argumentos:
            string (str): A string a ser inserida.
            index (int): O índice onde a string deve ser inserida.
        """
        if index > self.size:
            print("Erro! A posição desejada não existe na lista")
        else:
            newNode = Node(string)

            if self.isEmpty():
                self.head = newNode
                self.tail = newNode
            elif index == 0:
                newNode.next = self.head
                self.head.prev = newNode
                self.head = newNode
            elif index == self.size:
                newNode.prev = self.tail
                self.tail.next = newNode
                self.tail = newNode
            else:
                prevNode = self.head
                for _ in range(index - 1):
                    prevNode = prevNode.next

                newNode.next = prevNode.next
                newNode.prev = prevNode
                prevNode.next.prev = newNode
                prevNode.next = newNode

            self.size += 1

    def __str__(self):
        """
        Converte a lista em uma string.

        Retorna:
            Uma representação em string da lista.
        """
        node = self.head
        nodes = []
        while node is not None:
            nodes.append(node.data)
            node = node.next

        if nodes:
            return '[' + ', '.join(nodes[:-1]) + ']' + nodes[-1] + ', '
        else:
            return ''

##########################################################################

# testa o exercicio
lista = ListaBolha()        

# inserir no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar bola")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n\n", lista)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, Centro Universitario Facens, ', end="")
print('O menino gosta de andar, ', end="")
print('A menina gosta de jogar bola]', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")
````

Modifique a classe do exercício anterior modificando seu nome para **ListaBuscaBinaria**. Com a modificação do exercício anterior, é garantido que os elementos da lista sempre estarão ordenados. Você irá aproveitar esse situação e criar um método chamado **buscaValor** que aplica a busca binária ensinada na disciplina para verificar se o valor buscado existe ou não. Você deve implementar a versão recursiva da busca binária dentro desse método.  

````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class ListaBuscaBinaria(ListaBolha):
    def buscaValor(self, valor, inicio=None, fim=None):
        """
        Realiza uma busca binária na lista para encontrar um valor.

        Argumentos:
            valor (str): O valor a ser buscado.
            inicio (Node, opcional): O nó inicial para a busca. 
                                     Se None, a busca começa no início da lista.
            fim (Node, opcional): O nó final para a busca. 
                                   Se None, a busca termina no final da lista.

        Retorna:
            bool: True se o valor for encontrado, False caso contrário.
        """
        if inicio is None:
            inicio = self.head
        if fim is None:
            fim = self.tail

        if inicio is None or fim is None:
            return False

        meio = self.getMid(inicio, fim)

        if meio.data == valor:
            return True
        elif meio == inicio or meio == fim:
            return False
        elif meio.data < valor:
            return self.buscaValor(valor, meio.next, fim)
        else:
            return self.buscaValor(valor, inicio, meio.prev)

    def getMid(self, inicio, fim):
        """
        Obtém o nó do meio entre dois nós.

        Argumentos:
            inicio (Node): O nó inicial.
            fim (Node): O nó final.

        Retorna:
            Node: O nó do meio entre o início e o fim.
        """
        slow = inicio
        fast = inicio

        while fast is not fim and fast.next is not fim:
            slow = slow.next
            fast = fast.next.next

        return slow




##########################################################################

# testa o exercicio
lista = ListaBuscaBinaria()        

# inserir no inicio da lista
lista.insert("Centro Universitario Facens")
lista.insert("Eu fui para um shopping em Sorocaba")
lista.insert("A menina gosta de jogar bola")
lista.insert("O menino gosta de andar")
lista.insert("Tá tá tá")

print("\nLista:\n", lista)

encontrou = lista.buscaValor("O menino gosta de andar")
print("Encontrou?: ", encontrou)

encontrou = lista.buscaValor("O menino gosta de correr")
print("\nEncontrou?: ", encontrou)

print('\n' + 20*'-')
print('Resultado esperado:')
print("\nLista:\n")
print('[Tá tá tá, Centro Universitario Facens, ', end="")
print('O menino gosta de andar, ', end="")
print('A menina gosta de jogar bola]', end="")
print('Eu fui para um shopping em Sorocaba, ', end="")

print('\nEncontrou?: True')
print('\nEncontrou?: False')
````

## Exercício 6

Modifique a árvore ensinada nessa disciplina criando uma função que receba um valor como entrada, procure esse valor na árvore e retorne a quantidade de subnós (nós abaixo dele, incluindo ele mesmo). Se o nó não for encontrado, deve retornar 0. 


````python
################# ESCREVA SEU CÓDIGO AQUI  ###############################

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(data, self.root)

    def _insert(self, data, node):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert(data, node.left)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert(data, node.right)

    def contarSubnos(self, data):
        node = self.busca(data, self.root)
        if node is None:
            return 0
        else:
            return self.contaSubNos(node)

    def busca(self, data, node):
        if node is None or node.data == data:
            return node
        if data < node.data:
            return self.busca(data, node.left)
        else:
            return self.busca(data, node.right)

    def contaSubNos(self, node):
        count = 1
        if node.left is not None:
            count += self.contaSubNos(node.left)
        if node.right is not None:
            count += self.contaSubNos(node.right)
        return count



##########################################################################

tree1 = Tree()
tree1.insert(6)
tree1.insert(2)
tree1.insert(7)
tree1.insert(1)
tree1.insert(4)
tree1.insert(3)
tree1.insert(5)

print(tree1.contarSubnos(6))
print(tree1.contarSubnos(2))
print(tree1.contarSubnos(1))

print('\n' + 20*'-')
print('Resultado esperado:')
print("7")
print("5")
print("1")
````
